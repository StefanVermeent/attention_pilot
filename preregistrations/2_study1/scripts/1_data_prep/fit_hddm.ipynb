```python
import hddm
import pandas as pd

data = pd.read_csv('~/work/data/2_study1/hddm_data.csv')

hddm_data_enh = data[data['condition'] == 'enhanced']
hddm_data_deg = data[data['condition'] == 'degraded']
hddm_data_std = data[data['condition'] == 'standard']
```

```python
mod_fixed_enh = hddm.HDDM(data = hddm_data_enh, bias = False, depends_on = ({'v': 'congruency', 't': 'congruency'}))        
mod_fixed_enh.find_starting_values()           
hddm_fixed_enh = mod_fixed_enh.sample(100000, burn=50000, thin = 10, dbname='traces_enh.db', db='pickle') 
```

```python
mod_fixed_deg = hddm.HDDM(data = hddm_data_deg, bias = False, depends_on = ({'v': 'congruency', 't': 'congruency'}))        
mod_fixed_deg.find_starting_values()           
hddm_fixed_deg = mod_fixed_deg.sample(100000, burn=50000, thin = 10, dbname='traces_deg.db', db='pickle')
```

```python
mod_fixed_std = hddm.HDDM(data = hddm_data_std, bias = False, depends_on = ({'v': 'congruency', 't': 'congruency'}))        
mod_fixed_std.find_starting_values()           
hddm_fixed_std = mod_fixed_std.sample(100000, burn=50000, thin = 10, dbname='traces_std.db', db='pickle')
```

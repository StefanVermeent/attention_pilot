```python
# Libraries
import pandas as pd
import hddm

data = pd.read_csv("~/hddm/data/2_study1/hddm_data.csv")
iv_data = pd.read_csv("~/hddm/data/2_study1/2_cleaned_data.csv")
iv_data = iv_data[["id", "vio_comp", "unp_comp"]]
iv_data = iv_data.rename(columns={'id': 'subj_idx'})
iv_data

data = data.merge(iv_data)
data
```

```python
# 1. Standard Condition 

## 1.1 Parameters are fixed across congruency type

data_std = data[data.condition == 'standard']

# Set up model with constraint for the bias parameter
mod_fixed_std = hddm.HDDM(data_std, bias = False)

# find a good starting point which helps with the convergence.
mod_fixed_std.find_starting_values()
# start drawing 4000 samples and discarding 1000 as burn-in 
mod_fixed_std.sample(4000, burn=1000)

# Extract group-level/individual parameter estimates and posteriors
parms_fixed_std   = mod_fixed_std.gen_stats()
traces_fixed_std  = mod_fixed_std.get_traces()

mod_fixed_std.plot_posteriors()

## 1.2 Separate parameters for congruency conditions

## Set up model with constraint for the bias parameter
mod_free_std = hddm.HDDM(data_std, bias = False)

## find a good starting point which helps with the convergence.
mod_free_std.find_starting_values()
## start drawing 4000 samples and discarding 1000 as burn-in 
mod_free_std.sample(4000, burn=1000)
```

```python
## 1.2 Separate parameters for congruency conditions

## Set up model with constraint for the bias parameter
mod_free_std = hddm.HDDM(data_std, bias = False, depends_on={'v': 'congruency', 't': 'congruency'})
#
## find a good starting point which helps with the convergence.
mod_free_std.find_starting_values()
## start drawing 4000 samples and discarding 1000 as burn-in 
mod_free_std.sample(4000, burn=1000)
#
## Extract group-level/individual parameter estimates and posteriors
parms_free_std   = mod_free_std.gen_stats()
traces_free_std  = mod_free_std.get_traces()
```

```python
# This model does not converge

mod_free_std.plot_posteriors()
```
